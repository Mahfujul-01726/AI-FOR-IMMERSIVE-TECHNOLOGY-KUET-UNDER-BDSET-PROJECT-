```python
x=df['x']
y=df['y']
```

# Gradient Descent in machine learning

```python
# first step
m=0
c=0
Learning_rate=0.001

n= int(len(x))

iteration=1000

# second step
for i in range(iteration):
  y_prediction=m*x+c
  derivatives_m=(-2/n)+ sum(x*(y-y_prediction))
  derivatives_c=(-2/n)+ sum(y-y_prediction)

  # final work

  m= m- (Learning_rate + derivatives_m)
  c= c- (Learning_rate + derivatives_c)

print(m, c)
```
