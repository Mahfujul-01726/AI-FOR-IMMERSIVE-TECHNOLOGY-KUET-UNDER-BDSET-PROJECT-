```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

```

```python
# Generate synthetic signals
np.random.seed(0)
time = np.linspace(0, 10, 1000)

```

```python
# Original independent signals
s1 = np.sin(2 * time)  # Sinusoidal signal
s2 = np.sign(np.sin(3 * time))  # Square wave
S = np.c_[s1, s2]

```

```python
# Mix the signals
A = np.array([[1, 0.5], [0.5, 1]])  # Mixing matrix
X = S @ A.T  # Mixed signals


```

```python
# Apply ICA
ica = FastICA(n_components=2)
S_estimated = ica.fit_transform(X)  # Reconstructed signals
A_estimated = ica.mixing_  # Estimated mixing matrix

```

```python
# Plot results
plt.figure(figsize=(10, 6))
```

```python

# Plot original signals
plt.subplot(3, 1, 1)
plt.title("Original Signals")
plt.plot(S)
plt.grid()

```

```python
# Plot mixed signals
plt.subplot(3, 1, 2)
plt.title("Mixed Signals")
plt.plot(X)
plt.grid()

```

```python
# Plot separated signals
plt.subplot(3, 1, 3)
plt.title("Recovered Signals (ICA)")
plt.plot(S_estimated)
plt.grid()

plt.tight_layout()
plt.show()
```
