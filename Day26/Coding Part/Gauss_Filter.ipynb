```python
image_path = '/content/drive/MyDrive/Colab Notebooks/pic.png'
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Read the image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the input image
#cv2.imshow("Input Image", img)
cv2_imshow(img)

# Apply Gaussian blur
blur = cv2.GaussianBlur(img, (7, 7), 3, cv2.BORDER_DEFAULT)

# Display the blurred image
#cv2.imshow('Output Image (Blurred)', blur)
cv2_imshow(blur)

# Save the blurred image to a file
#cv2.imwrite('blurred_image.png', blur)
#print("Blurred image saved as 'blurred_image.png'")

# Wait for a key press and close all windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
